## Data Cleaning Assignment 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
              'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
              'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
df['FlightNumber'].interpolate(inplace=True)
```

```python
df['FlightNumber']=df.FlightNumber.astype(int)
```

```python
df.head()
```

```python
df1 = df['From_To'].str.split("_",expand=True,)
```

```python
df.head()
```

```python
df1.head()
```

```python
df1.rename(columns = {0:'First',1:'Last'}, inplace=True)
```

```python
df1['Last']  = df1['Last'].str.capitalize()
```

```python
df1['First']  = df1['First'].str.capitalize()
```

```python
df1
```

```python
df.drop('From_To', axis=1, inplace=True)
```

```python
df.head()
```

```python
df2=df1.join(df)
```

```python
df2
```

```python
delays = df2['RecentDelays'].apply(pd.Series)
```

```python
delays = delays.rename(columns = lambda x: 'delays_' + str(x))
```

```python
delays
```

```python
df2.drop('RecentDelays',axis=1, inplace=True)
df2
```

```python
df3 = pd.concat([df2[:],delays[:]],axis=1)
```

```python
df3
```

```python
df3.insert(6, 'Airline', df3.pop('Airline'))
```
